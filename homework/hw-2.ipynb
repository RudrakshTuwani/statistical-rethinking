#### 2M1

```python
grid_size = 50

# Grid search values
p_grid = np.linspace(0, 1, num=grid_size)

# Set priors
priors = np.array([1]*grid_size)

def calc_likelihood(data, p_grid, success):
    k = Counter(data)[success]
    n = len(data)
    
    return binom.pmf(k=k, n=n, p=p_grid)

def calc_posterior(data, p_grid, priors, success):
    likelihood = calc_likelihood(data, p_grid, success)
    num = priors * likelihood
    posterior = num / sum(num)
    
    return posterior

def plot(p_grid, posterior_prob, **args):
    plt.plot(p_grid, posterior_prob, marker='o', **args);
    plt.xlabel('Probability of water')
    plt.ylabel('Density')
```

##### 1. W,W,W

```python
data = ['W', 'W', 'W']
posterior_prob = calc_posterior(data, p_grid, priors, 'W')
plot(p_grid, posterior_prob)
```

##### 2. W,W,W,L

```python
data = ['W', 'W', 'W', 'L']
posterior_prob = calc_posterior(data, p_grid, priors, 'W')
plot(p_grid, posterior_prob)
```

##### 3. L,W,W,L,W,W,W

```python
data = ['L', 'W', 'W', 'L', 'W', 'W', 'W'] 
posterior_prob = calc_posterior(data, p_grid, priors, 'W')
plot(p_grid, posterior_prob)
```

```python
# Prior probability of choosing
prior_earth = 1
prior_mars = 1

# Likelihood of getting land
likelihood_earth = binom.pmf(1, 1, p=0.3)
likelihood_mars = binom.pmf(1, 1, p=1.0)

# Posterior prob of getting Earth given that land observed
num_earth = likelihood_earth * prior_earth
num_mars = likelihood_mars * prior_mars

posterior_earth_given_land = num_earth / (num_earth + num_mars)
print(posterior_earth_given_land)
```
